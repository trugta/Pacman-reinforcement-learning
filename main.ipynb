```python
import gymnasium as gym
import ale_py
import pygame
import time
```

```python

# Đăng ký môi trường Atari với Gymnasium
gym.register_envs(ale_py)

# Khởi tạo môi trường Pacman thay vì CartPole
env = gym.make('ALE/Pacman-v5', render_mode="human")

# Khởi tạo lại môi trường cho episode mới
observation, info = env.reset()

print(f"Khởi tạo thành công Pacman! Kích thước ảnh: {observation.shape}")

# SỬA LỖI 2: Điền giá trị khởi tạo cho biến vòng lặp
episode_over = False  
total_reward = 0

while not episode_over:
    # Bắt sự kiện bấm nút [X] để đóng cửa sổ an toàn
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            episode_over = True
            break
            
    if episode_over:
        break
        
    # Chọn hành động ngẫu nhiên (Pacman có nhiều hành động hơn CartPole)
    action = env.action_space.sample()  

    # Thực hiện hành động
    observation, reward, terminated, truncated, info = env.step(action)

    total_reward += reward
    episode_over = terminated or truncated
    
    # SỬA LỖI 3: Thêm delay nhỏ để tránh game chạy quá nhanh không kịp nhìn
    time.sleep(0.01)  

print(f"Trận đấu kết thúc! Tổng điểm đạt được: {total_reward}")

# Đóng cửa sổ và giải phóng bộ nhớ đồ họa hoàn toàn
env.close()
pygame.quit()
```
